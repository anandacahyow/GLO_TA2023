```python
import numpy as np
from scipy import signal
import pandas as pd

df = pd.read_csv("welltest.csv")
#df = pd.read_csv("well_VX2.csv")
x1 = df['glir'].to_numpy()
x = signal.resample(x1, len(x1) * 24)
x_int = x.astype(int)

y1 = df['qo'].to_numpy()
y = signal.resample(y1, len(y1) * 24)
y_int = y.astype(int)
```

```python
from matplotlib import pyplot as plt
xx = df['glir'].to_numpy()
yy = df['qo'].to_numpy()

xxx = signal.resample_poly(xx, len(xx) * 5, len(xx) * 1)
yyy = signal.resample_poly(yy, len(yy) * 5, len(yy) * 1)

t = range(0, len(xxx))
np.shape(t)
plt.plot(t,xxx,label='glir')
plt.plot(t,yyy,label='qo')
plt.legend()
plt.grid()

print(xxx,'\n\n',yyy)
print(len(xxx),'\n\n',len(yyy))
```
